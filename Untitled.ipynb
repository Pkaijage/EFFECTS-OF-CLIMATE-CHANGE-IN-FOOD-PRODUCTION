```python
#for setting width and height.
plt.figure(figsize=(16, 10))

#visuals plots
db.plot.bar()
```

```python
#pie chart
plt.figure(figsize=(7, 5))
count_per_label = [len(db[db['CL'] == 0]),len(db[db['CL'] == 1])]
labels = [0, 1]
colors = ['yellowgreen', 'lightblue']
explode = (0.05, 0.1)
plt.pie(count_per_label, explode=explode, labels=labels, 
        colors=colors,autopct='%4.2f%%',shadow=True, startangle=45)
plt.title('CASSAVA LEVEL DISTRIBUTION')
plt.axis('equal')
plt.show()
```
